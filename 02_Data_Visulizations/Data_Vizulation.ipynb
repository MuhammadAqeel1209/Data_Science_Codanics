# Data Visualization

```python
# Import Libiray
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data Titanic from Seaborn Libirary 
data = sns.load_dataset("titanic")
```

```python
data.plot(kind='scatter',x='age',y='fare')
```

```python
data.plot(kind='scatter',x='age',y='fare',color = "#FE7F0E")
```

```python
data.groupby('sex')['age'].mean().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Age")
```

```python
# Bar plot using seaborn 
plt.figure(figsize=(10,6))
sns.barplot(x='survived', y='fare', data=data)
```

```python
sns.scatterplot(x='fare', y='age', data=data,hue='class')
```

```python
sns.scatterplot(x='fare', y='age', data=data,hue='sex')
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder()
px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
```
