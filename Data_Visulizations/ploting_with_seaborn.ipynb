Importing Libiraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

#### Important things for make graph

- X axis
- Y axis

```python
# Load the data set
df = sns.load_dataset('titanic')
```

```python
df.info()
```

# Scatter Plot

```python
# Scatter PLot
plt.figure(figsize = (10,8))
# Change the Color 
# Define the colors for each class
palette = {'First': 'red', 'Second': 'blue', 'Third': 'yellow'}
sns.scatterplot(data = df, x='age',y='fare',hue='class',palette=palette)

# Label
plt.xlabel('age')
plt.ylabel('fare')
plt.title("Age and Fare Based on Class")

# Saving a Plot
# plt.savefig('age_fare_class.png',dpi = 300)
# plt.savefig('age_fare_class.svg')
```

```python
# sns.set_style('darkgrid')

sns.set_theme(style='darkgrid',font_scale = 1.3,font = "Times New Roman")

plt.figure(figsize = (10,8))
# Change the Color 
# Define the colors for each class
palette = {'male': 'red', 'female': 'green'}
sns.scatterplot(data = df, x='age',y='fare',hue='sex',palette=palette)

# Label
plt.xlabel('age')
plt.ylabel('fare')
plt.title("Age and Fare Based on Gender")
```

##### Interperatation 
- Age of Passenger at X Axis and Fare at Y Axis
- The Scatter plot show the relationship between age and fare of titanic passenger 
- We further grouped the passenger by their Gender male in red points and female in green points
- The Oldest person in titanic is male having 80 year old as show in scatter plot 

# Box PLot

```python
# sns.set_style('darkgrid')

sns.set_theme(style='darkgrid',font_scale = 1.3,font = "Times New Roman")

plt.figure(figsize = (10,8))
# Change the Color 
# Define the colors for each class
palette = {'First': 'red', 'Second': 'blue', 'Third': 'yellow'}
sns.boxplot(data = df, x='sex',y='age',hue='class',palette=palette)

# Label
plt.xlabel('gender')
plt.ylabel('age')
plt.title("Gender and Age Based on Class")

# Change the title and position of legend
plt.legend(title = "Class", loc='upper right')
```

# Boxen PLot

```python
# sns.set_style('darkgrid')

sns.set_theme(style='darkgrid',font_scale = 1.3,font = "Times New Roman")

plt.figure(figsize = (10,8))
# Change the Color 
# Define the colors for each class
palette = {'First': 'red', 'Second': 'blue', 'Third': 'yellow'}
sns.boxenplot(data = df, x='sex',y='age',hue='class',palette=palette)

# Label
plt.xlabel('gender')
plt.ylabel('age')
plt.title("Gender and Age Based on Class")

# Change the title and position of legend
plt.legend(title = "Class", loc='upper right')
```

# Bar Plot

```python
# sns.set_style('darkgrid')

sns.set_theme(style='darkgrid',font_scale = 1.3,font = "Times New Roman")

plt.figure(figsize = (10,8))
# Change the Color 
# Define the colors for each class
palette = {'First': 'red', 'Second': 'blue', 'Third': 'yellow'}
sns.barplot(data = df, x='sex',y='age',hue='class',palette=palette,capsize=0.1)

# Label
plt.xlabel('gender')
plt.ylabel('age')
plt.title("Gender and Age Based on Class")

# Change the title and position of legend
plt.legend(title = "Class", loc='upper right')
```

```python
# sns.set_style('darkgrid')

sns.set_theme(style='darkgrid',font = "Times New Roman")

plt.figure(figsize = (8,4))
# Change the Color 
# Define the colors for each class
palette = {'male': 'red', 'female': 'green'}
sns.scatterplot(data = df, x='age',y='fare',hue='sex',size = 'age',palette=palette)

# Label
plt.xlabel('age')
plt.ylabel('fare')
plt.title("Age and Fare Based on Gender")
```
